## Add Two Numbers

**LeetCode Link:** https://leetcode.com/problems/add-two-numbers/

---

## Problem Statement

You are given two non-empty linked lists representing two non-negative integers.  
The digits are stored in **reverse order**, and each node contains a single digit.

Add the two numbers and return the sum as a linked list.

You may assume:
- The two numbers do not contain leading zeros (except the number 0 itself).
- Each nodeâ€™s value is between 0 and 9.



```python
# Helper class and functions for testing
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def build_linked_list(arr):
    dummy = ListNode(0)
    curr = dummy
    for val in arr:
        curr.next = ListNode(val)
        curr = curr.next
    return dummy.next

def print_linked_list(head):
    res = []
    while head:
        res.append(head.val)
        head = head.next
    return res
```

## Solution

```python
class Solution(object):
    def addTwoNumbers(self, l1, l2):
        dummy = ListNode(0)
        curr = dummy
        carry = 0

        while l1 or l2 or carry:
            val1 = l1.val if l1 else 0
            val2 = l2.val if l2 else 0

            total = val1 + val2 + carry
            carry = total // 10
            digit = total % 10

            curr.next = ListNode(digit)
            curr = curr.next

            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next

        return dummy.next
```

```python
l1 = build_linked_list([2, 4, 3])
l2 = build_linked_list([5, 6, 4])
print(print_linked_list(Solution().addTwoNumbers(l1, l2)))  # [7, 0, 8]

# Test Case 2
l1 = build_linked_list([0])
l2 = build_linked_list([0])
print(print_linked_list(Solution().addTwoNumbers(l1, l2)))  # [0]

# Test Case 3
l1 = build_linked_list([9,9,9,9,9,9,9])
l2 = build_linked_list([9,9,9,9])
print(print_linked_list(Solution().addTwoNumbers(l1, l2)))  # [8,9,9,9,0,0,0,1]
```

### Explanation
This solution performs addition directly on the linked lists, similar to how we add numbers by hand.

Steps:
1. Use a dummy node to simplify result list construction.
2. Traverse both lists simultaneously.
3. Add corresponding digits along with a carry.
4. Store the result digit in a new node.
5. Continue until both lists and the carry are exhausted.

### Complexity Analysis 

- **Time Complexity:** O(max(n, m))  
- **Space Complexity:** O(max(n, m))
