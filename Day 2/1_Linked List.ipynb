Why Do Linked Lists Exist?

Because arrays are bad at some things.

Arrays are good at:

Fast access: O(1)

Simple indexing

Arrays are bad at:

Insertions/deletions in the middle → O(n)

Resizing costs

Linked Lists are good at:

Insertions/deletions → O(1) if you have the node

Dynamic size (no resizing)

Linked Lists are bad at:

Accessing elements → O(n)

More memory (extra pointers)
