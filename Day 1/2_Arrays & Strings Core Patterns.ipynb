## Arrays & Strings: Core Patterns

These patterns appear in **~80% of junior-level coding interviews**.  
The goal is **recognition**, not mastery (yet).

**Used when:**
- Checking duplicates
- Comparing frequencies
- Matching pairs

**Typical questions:**
- Are two strings anagrams?
- Does an array contain duplicates?
- How many times does each element appear?

**Why it works:**
- Hash maps provide **O(1)** average lookup time

### ðŸ” Pattern 2: Two Pointers

**Used when:**
- Reversing a string or array
- Checking palindromes
- Working with **sorted arrays**

**Typical questions:**
- Is this string a palindrome?
- Merge two sorted arrays
- Remove duplicates in place

**Why it works:**
- Reduces extra space usage
- Often improves time complexity to **O(n)**

---

### ðŸ” Pattern 3: Sliding Window

**Used when:**
- Finding the **longest** or **shortest** substring
- Working with **subarrays**
- Handling **continuous ranges**

**Typical questions:**
- Longest substring without repeating characters
- Maximum sum subarray of size `k`

**Why it works:**
- Avoids recalculating overlapping ranges
- Maintains an efficient **moving window**
